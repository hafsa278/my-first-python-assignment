```python
# assignment no 12
password = input("enter your password")
upperCase = 0
lowerCase = 0
digit = 0
for i in password:
  if i.isupper():
      upperCase+=1
  elif i.islower():
     lowerCase+=1
  elif i.isdigit():
       digit+=1
print(f" your password has {digit} character. {upperCase} upperCase and {lowerCase}  lowerCase")
if len(password) <6 and upperCase == 0 and digit == 0:
  print("your password is weeak")
elif len(password) >=6 and upperCase>=1 and digit >= 1:
    print("your password is medium")
elif len(password) >=8 and upperCase >= 2 and digit>=2:
    print("your password is strong")
```
